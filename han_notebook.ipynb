### 3. Filter rows based on conditions
As observed from the data above, there are many missing values in key columns that indicate the severity of the disasters. I aim to identify the different types of disasters Japan has experienced, such as earthquakes, storms, floods, etc., and calculate the average value for each category to fill in the missing data. Using the mean is appropriate since the average will be calculated for each disaster type individually, rather than across all disaster types combined. First is to find all the disasters types that Japan was involved in then filter based on those categories to calculate average for each disaster type. As a result you will see **22** different types of disasters that happened in Japan

```python
import pandas as pd

# CHANGE csvFilename as necessary
csvFilename = 'emdat_processed.csv'

# CHANGE these for your specified search
column = 'Country'
value = 'Japan'
specifiedColumns = ['Year', 'Country', 'Disaster Type']

csvDirectory = 'dataset_processed/{}'.format(csvFilename)
data = pd.read_csv(csvDirectory)

def readData():
    # should return about 382 rows of disasters that happened in JAPAN.
    # using this, we need to find the disaster types
    japanData = data[(data['{}'.format(column)] == '{}'.format(value))][specifiedColumns]
    
    return japanData['Disaster Type'].drop_duplicates()

readData()
```

### 4. Finding the average Total Damage, Total Affected, and Total Deaths based on average of available data
Using the information from previous return, we will find the average of each disaster type by using the mean function from pandas. Now that each of the disasters are defined, we will separate the dataframe by each disaster type and get the average value for each desired columns such as _Total Damage ('000 US$), Total Affected, and Total Deaths_. After the average values have been found, average values will be assigned to approriate area with a null value.

```python
import pandas as pd

# CHANGE csvFilename as necessary
csvFilename = 'emdat_processed.csv'

# CHANGE these for your specified search
column = 'Country'
value = 'Japan'
specifiedColumns = ['Year', 'Country', 'Disaster Type', "Total Damage ('000 US$)", 'Total Affected', 'Total Deaths']

csvDirectory = 'dataset_processed/{}'.format(csvFilename)
data = pd.read_csv(csvDirectory)

def readData():
    japanData = data[(data['{}'.format(column)] == '{}'.format(value))][specifiedColumns]
    return japanData['Disaster Type'].drop_duplicates()

def findAverage():
    noDuplicate = readData()
    japanData = data[(data['{}'.format(column)] == '{}'.format(value))][specifiedColumns]
    # iterate through each disaster type and find the average for total damage, total affected and total deaths
    for row in noDuplicate.T.items():
        disasterSpecificData = japanData[(japanData['Disaster Type'] == row[1])]
        # get the average; if all values are NaN, then simply assign 0 
        totalDamage = int(disasterSpecificData["Total Damage ('000 US$)"].mean()) if disasterSpecificData["Total Damage ('000 US$)"].any() else 0
        totalAffected = int(disasterSpecificData["Total Affected"].mean()) if disasterSpecificData["Total Affected"].any() else 0
        totalDeaths = int(disasterSpecificData["Total Deaths"].mean()) if disasterSpecificData["Total Deaths"].any() else 0
        # assign to the NaN values
        japanData.loc[(japanData['Disaster Type'] == row[1]) & disasterSpecificData["Total Damage ('000 US$)"].isna(), "Total Damage ('000 US$)"] = totalDamage
        japanData.loc[(japanData['Disaster Type'] == row[1]) & disasterSpecificData["Total Affected"].isna(), "Total Affected"] = totalAffected
        japanData.loc[(japanData['Disaster Type'] == row[1]) & disasterSpecificData["Total Deaths"].isna(), "Total Deaths"] = totalDeaths
    return japanData
        
findAverage()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


# CHANGE csvFilename as necessary
csvFilename = 'emdat_processed.csv'

# CHANGE these for your specified search
column = 'Country'
value = 'Japan'
specifiedColumns = ['Year', 'Country', 'Disaster Type', "Total Damage ('000 US$)", 'Total Affected', 'Total Deaths']

csvDirectory = 'dataset_processed/{}'.format(csvFilename)
data = pd.read_csv(csvDirectory)

def readData():
    japanData = data[(data['{}'.format(column)] == '{}'.format(value))][specifiedColumns]
    return japanData['Disaster Type'].drop_duplicates()

def findAverage():
    noDuplicate = readData()
    japanData = data[(data['{}'.format(column)] == '{}'.format(value))][specifiedColumns]
    # iterate through each disaster type and find the average for total damage, total affected and total deaths
    for row in noDuplicate.T.items():
        disasterSpecificData = japanData[(japanData['Disaster Type'] == row[1])]
        # get the average; if all values are NaN, then simply assign 0 
        totalDamage = int(disasterSpecificData["Total Damage ('000 US$)"].mean()) if disasterSpecificData["Total Damage ('000 US$)"].any() else 0
        totalAffected = int(disasterSpecificData["Total Affected"].mean()) if disasterSpecificData["Total Affected"].any() else 0
        totalDeaths = int(disasterSpecificData["Total Deaths"].mean()) if disasterSpecificData["Total Deaths"].any() else 0
        # assign to the NaN values
        japanData.loc[(japanData['Disaster Type'] == row[1]) & disasterSpecificData["Total Damage ('000 US$)"].isna(), "Total Damage ('000 US$)"] = totalDamage
        japanData.loc[(japanData['Disaster Type'] == row[1]) & disasterSpecificData["Total Affected"].isna(), "Total Affected"] = totalAffected
        japanData.loc[(japanData['Disaster Type'] == row[1]) & disasterSpecificData["Total Deaths"].isna(), "Total Deaths"] = totalDeaths
    return japanData

japanData = findAverage()

def plot_data(japanData):
    eqData = japanData[(japanData['Disaster Type'] == 'Earthquake')]
    plt.figure(figsize=(20, 10))
    
    plt.subplot(3, 1, 1)
    plt.xlabel('Year')
    plt.ylabel('Total Damage ($)')
    plt.plot(eqData['Year'], eqData["Total Damage ('000 US$)"], color='b')
    plt.xticks(eqData['Year'].unique(), rotation=45)
    plt.title("JAPAN: Earthquake - Total Damage ('000 US$)")
    
    plt.subplot(3, 1, 2)
    plt.xlabel('Year')
    plt.ylabel('Total Affected')
    plt.plot(eqData['Year'], eqData["Total Affected"], color='b')
    plt.xticks(eqData['Year'].unique(), rotation=45)
    plt.title("JAPAN: Earthquake - Total Affected")
    
    plt.subplot(3, 1, 3)
    plt.xlabel('Year')
    plt.ylabel('Total Deaths')
    plt.plot(eqData['Year'], eqData["Total Deaths"], color='b')
    plt.xticks(eqData['Year'].unique(), rotation=45)
    plt.title("JAPAN: Earthquake - Total Deaths")
    
    plt.tight_layout()
    plt.show()

plot_data(japanData)
```

### 6. Make correlation with world_bank data
Now that we have an understanding of the damage caused by earthquakes in Japan over the years, we will compare this data with Japan's economic indicators from the world_bank database. This comparison will help identify potential correlations between natural disasters and economic performance. For this example, we will compare the indicator _Electronic power consumption_ from the world_bank dataset to see if there is any correlation between earthquakes with high damage with electronic power consumption.
```python
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.ticker as ticker


# CHANGE csvFilename as necessary
csvFilename = 'emdat_processed.csv'
csvFilename2 = 'world_bank_processed.csv'

# CHANGE these for your specified search
column = 'Country'
value = 'Japan'
specifiedColumns = ['Year', 'Country', 'Disaster Type', "Total Damage ('000 US$)", 'Total Affected', 'Total Deaths']

csvDirectory = 'dataset_processed/{}'.format(csvFilename)
data = pd.read_csv(csvDirectory)

csvDirectory2 = 'dataset_processed/{}'.format(csvFilename2)
data2 = pd.read_csv(csvDirectory2)

def readData():
    japanData = data[(data['{}'.format(column)] == '{}'.format(value))][specifiedColumns]
    return japanData['Disaster Type'].drop_duplicates()

def findAverage():
    noDuplicate = readData()
    japanData = data[(data['{}'.format(column)] == '{}'.format(value))][specifiedColumns]
    # iterate through each disaster type and find the average for total damage, total affected and total deaths
    for row in noDuplicate.T.items():
        disasterSpecificData = japanData[(japanData['Disaster Type'] == row[1])]
        # get the average; if all values are NaN, then simply assign 0 
        totalDamage = int(disasterSpecificData["Total Damage ('000 US$)"].mean()) if disasterSpecificData["Total Damage ('000 US$)"].any() else 0
        totalAffected = int(disasterSpecificData["Total Affected"].mean()) if disasterSpecificData["Total Affected"].any() else 0
        totalDeaths = int(disasterSpecificData["Total Deaths"].mean()) if disasterSpecificData["Total Deaths"].any() else 0
        # assign to the NaN values
        japanData.loc[(japanData['Disaster Type'] == row[1]) & disasterSpecificData["Total Damage ('000 US$)"].isna(), "Total Damage ('000 US$)"] = totalDamage
        japanData.loc[(japanData['Disaster Type'] == row[1]) & disasterSpecificData["Total Affected"].isna(), "Total Affected"] = totalAffected
        japanData.loc[(japanData['Disaster Type'] == row[1]) & disasterSpecificData["Total Deaths"].isna(), "Total Deaths"] = totalDeaths
    return japanData

japanData = findAverage()

def correlationData():
    # years = [str(year) for year in range(1960, 2024)]
    # df_cleaned = japanEcoData.dropna(subset=years, how='any')
    # return df_cleaned
    return data2[(data2['Country Name'] == 'Japan') & (data2['Indicator Name'] == 'Electric power consumption (kWh per capita)')]

def plot_data(japanData):
    eqData = japanData[(japanData['Disaster Type'] == 'Earthquake')]
    ecoData = correlationData()
    years = [str(year) for year in range(1960, 2024)]
    
    plt.figure(figsize=(20, 30))
    
    plt.subplot(4, 1, 1)
    plt.xlabel('Year')
    plt.ylabel('Total Damage ($)')
    plt.plot(eqData['Year'], eqData["Total Damage ('000 US$)"], color='b')
    plt.xticks(eqData['Year'].unique(), rotation=45)
    plt.title("JAPAN: Earthquake - Total Damage ('000 US$)")
    
    plt.subplot(4, 1, 2)
    plt.xlabel('Year')
    plt.ylabel('Total Affected')
    plt.plot(eqData['Year'], eqData["Total Affected"], color='b')
    plt.xticks(eqData['Year'].unique(), rotation=45)
    plt.title("JAPAN: Earthquake - Total Affected")
    
    plt.subplot(4, 1, 3)
    plt.xlabel('Year')
    plt.ylabel('Total Deaths')
    plt.plot(eqData['Year'], eqData["Total Deaths"], color='b')
    plt.xticks(eqData['Year'].unique(), rotation=45)
    plt.title("JAPAN: Earthquake - Total Deaths")
    
    plt.subplot(4, 1, 4)
    plt.xlabel('Year')
    plt.ylabel('Electric power consumption (kWh per capita)')
    for year in years:
        plt.plot(year, ecoData[year], '-o', markersize=5, color = 'r')
    plt.xticks(rotation=45)
    plt.title("JAPAN: Electric power consumption (kWh per capita)")
    
    
    plt.tight_layout()
    plt.show()

plot_data(japanData)
```
