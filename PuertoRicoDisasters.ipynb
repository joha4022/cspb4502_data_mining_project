```python
# Return indicators with 10 or less NaN values

nan_counts = PR_econ.isna().sum(axis=1)
PR_econ.loc[:, 'NaN Count'] = nan_counts
PR_econ_indicator = PR_econ[PR_econ["NaN Count"] <= 10]
PR_econ_indicator = PR_econ_indicator.drop(columns=['NaN Count'])
```

```python
for i in PR_econ_indicator['Indicator Name'].unique():
    print(i)
    
# I will use GDP growth (annual %)
```

```python
print(PR_df["Disaster Type"].unique())
storm = PR_df[PR_df["Disaster Type"] == 'Storm']
flood = PR_df[PR_df["Disaster Type"] == 'Flood']
chem_spill = PR_df[PR_df["Disaster Type"] == 'Chemical spill']
wet_mass_mov = PR_df[PR_df["Disaster Type"] == 'Mass movement (wet)']
fire = PR_df[PR_df["Disaster Type"] == 'Fire (Miscellaneous)']
water = PR_df[PR_df["Disaster Type"] == 'Water']
drought = PR_df[PR_df["Disaster Type"] == 'Drought']
explosion_misc = PR_df[PR_df["Disaster Type"] == 'Explosion (Miscellaneous)']
air = PR_df[PR_df["Disaster Type"] == 'Air']
explosion_indust = PR_df[PR_df["Disaster Type"] == 'Explosion (Industrial)']
earthquake = PR_df[PR_df["Disaster Type"] == 'Earthquake']
```

```python
indicator_name = "GDP growth (annual %)"
indicator_data = PR_econ_indicator[PR_econ_indicator['Indicator Name'] == indicator_name]

# Extract year columns
years = PR_econ_indicator.columns[4:]  # Assuming the first 4 columns are non-year identifiers
values = indicator_data.iloc[0, 4:]  # Values for the specific indicator

# Convert year and values to numeric, dropping NaNs
years = years.astype(str)  # Convert years to string for labeling
values = values.dropna().values  # Drop NaN values for plotting

# Plotting
plt.figure(figsize=(12, 6))  # Increase the figure size
plt.plot(years[:len(values)], values, marker='o')
plt.title("Puerto Rico Disaster: Storms - GDP growth (annual %)")
plt.xlabel('Year')
plt.ylabel('Indicator Value')

disaster_years = storm["Year"] # Check Storms

# Add vertical lines for disaster years
for year in disaster_years:
    plt.axvline(x=str(year), color='red', linestyle='--')

# Adjust x-axis ticks
plt.xticks(ticks=range(0, len(years), 2), labels=years[::2], rotation=45, ha='right')  # Show every other year
plt.grid()
plt.tight_layout(pad=3.0)  # Adjust layout
plt.show()
```

```python
# Plotting
plt.figure(figsize=(12, 6))  # Increase the figure size
plt.plot(years[:len(values)], values, marker='o')
plt.title("Puerto Rico Disaster: Floods - GDP growth (annual %)")
plt.xlabel('Year')
plt.ylabel('Indicator Value')

disaster_years = flood["Year"] # Check Storms

# Add vertical lines for disaster years
for year in disaster_years:
    plt.axvline(x=str(year), color='orange', linestyle='--')

# Adjust x-axis ticks
plt.xticks(ticks=range(0, len(years), 2), labels=years[::2], rotation=45, ha='right')  # Show every other year
plt.grid()
plt.tight_layout(pad=3.0)  # Adjust layout
plt.show()
```

```python
# Plotting
plt.figure(figsize=(12, 6))  # Increase the figure size
plt.plot(years[:len(values)], values, marker='o')
plt.title("Puerto Rico Disaster: Earthquake - GDP growth (annual %)")
plt.xlabel('Year')
plt.ylabel('Indicator Value')

disaster_years = earthquake["Year"] # Check Storms

# Add vertical lines for disaster years
for year in disaster_years:
    plt.axvline(x=str(year), color='green', linestyle='--')

# Adjust x-axis ticks
plt.xticks(ticks=range(0, len(years), 2), labels=years[::2], rotation=45, ha='right')  # Show every other year
plt.grid()
plt.tight_layout(pad=3.0)  # Adjust layout
plt.show()
```

```python
# Group by "Year" and count disasters
grouped_data = PR_df[["Year", "Disaster Type"]].groupby("Year").agg(Count=('Disaster Type', 'count')).reset_index()

# Plotting
plt.figure(figsize=(12, 6))
sns.barplot(x='Year', y='Count', data=grouped_data, color='skyblue')
plt.title('Count of Disasters per Year')
plt.xlabel('Year')
plt.ylabel('Count of Disasters')
plt.xticks(rotation=45)
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```
